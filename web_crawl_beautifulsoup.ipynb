```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://en.wikipedia.org/wiki/A.J.W._McNeilly')
```

```python
html = response.text
```

```python
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup.title
```

```python
soup.find(id='mw-content-text').find(class_="mw-parser-output").p.a.get('href')
```
